## Data Preprocessing  

```python
data.head()
```

```python
df = pd.get_dummies(data['workclass'],drop_first=True)
df2= pd.get_dummies(data['education'],drop_first=True)
df1= pd.get_dummies(data['marital-status'],drop_first=True)
dff1= pd.get_dummies(data['occupation'],drop_first=True)
df3 = pd.get_dummies(data['relationship'],drop_first=True)
df4 = pd.get_dummies(data['race'],drop_first=True)
df5= pd.get_dummies(data['gender'],drop_first=True)
df6 = pd.get_dummies(data['native-country'],drop_first=True)
```

```python
data2 = pd.concat([df,df2,df1,dff1,df3,df4,df5,df6],axis=1)
```

```python
x={'<=50K':0,'>50K':1}
data['income']=data['income'].map(x)
```

```python
data.drop(columns=['workclass','education','marital-status','occupation','relationship','race','gender','native-country'],inplace=True)
```

```python
dat = pd.concat([data2,data],axis=1)
```

```python
len(dat.columns)
```

```python
x=dat.drop(columns=['income'])
y = dat['income']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.80,stratify=y,random_state=49)
```

```python
x_train.shape,y_train.shape
```

```python
scaler= StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
print('Accuarcy :',accuracy_score(y_test,y_pred))
```

## Select the Optimal **K** value

```python
score =[]

for i in range(1,100):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,Y_train)
    Y_pred = model.predict(X_test)
    scaores = accuracy_score(Y_test,Y_pred)
    score.append(scaores)
```

```python
np.array(score).max()
```

```python
plt.figure(figsize=(20,8))
plt.plot(range(1,100),score)
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
val =[]
for i in range(1,50):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,Y_train)
    Y_pred = model.predict(X_test)
    cv_score = cross_val_score(model,X,Y,cv=10,scoring='accuracy')
    val.append(cv_score.mean())
```

```python
val
```

```python
MSE=[1-x for x in val]
MSE
```

```python
Neibor = list(range(1,50))
k_optimal = Neibor[MSE.index(min(MSE))]
k_optimal
```

```python
minima=np.array(MSE).min()
minima
```
