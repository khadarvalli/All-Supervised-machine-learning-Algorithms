```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
x = datacorr.drop(columns=['class'])
y = datacorr['class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.80,random_state=0,stratify=y)
```

```python
x_train.shape,y_train.shape
```

```python
y_train = y_train.to_numpy()
x_train = x_train.to_numpy()
```

```python
print('y_train having 1 :-',len(y_train[y_train==1]))
print('y_train having o :-',len(y_train[y_train==0]))
```

```python
from imblearn.over_sampling import SMOTE 
```

```python
sm = SMOTE(sampling_strategy='minority',random_state=1,k_neighbors=5)
#if it is binary then we can use the 1 but the mupliple clssfication we use the 'minority'.
```

```python
x_train_res,y_train_res = sm.fit_sample(x_train,y_train.ravel())
```

```python
print('y_train having 1 :-',len(y_train_res[y_train_res==1]))
print('y_train having o :-',len(y_train_res[y_train_res==0]))
```

```python
model = DecisionTreeClassifier(criterion='gini',random_state=42)
model.fit(x_train_res,y_train_res)
y_pred = model.predict(x_test)
```

```python
print('Accuarcy :-',accuracy_score(y_test,y_pred))
```

```python
print('Classification Report :',classification_report(y_test,y_pred))
```
